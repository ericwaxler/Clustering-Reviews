```python
import pickle
import pprint

pp = pprint.PrettyPrinter(indent=4).pprint

[data, cols] = pickle.load(open('C:\\Users\\tom.lappas\\code\\travel\\data\\ratings.pkl','rb'))
```

What attraction types are the most popular? Even though these are averages, if someone skips a category while traveling it's going to be zero. And there are a lot of zeros in this dataset.

```python
for i, inst in enumerate(data):
    data[i] = [1 if x != 0 else 0 for x in inst]

sums = [sum(x) for x in zip(*data)]
    
print('Number of instances: {}\n'.format(len(data)))    
    
for i, col in enumerate(cols):
    print('{}: {}'.format(col, sums[i]))
```

I'm also curious if there are a small-ish number of patterns (0/1 for attributes) across instances.

Answer: Yes! Although I'm not sure if I'm surprized at how many or how few unique patterns are here.

```python
uniques = {}

for inst in data:
    key = tuple(inst)
    if key not in uniques:
        uniques[key] = 1
    else:
        uniques[key] = uniques[key] + 1
        
print('Number of unique activity patterns: {}\n'.format(len(uniques.keys())))

pp(uniques)
```

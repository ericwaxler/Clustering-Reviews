```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import pprint

pp = pprint.PrettyPrinter(indent=4).pprint
```

```python
[data, col_names] = pickle.load(open('C:\\Users\\tom.lappas\\code\\travel\\data\\ratings.pkl','rb'))

print('Number of instances: {}'.format(len(data)))
print('Example Instance:')
pp(data[0])

print('\nColumn Labels: ')
pp(col_names)
```

```python
rows = 5
cols = 5
n_plots = len(col_names)

fig, ax = plt.subplots(5, 5, sharey='row', figsize=(20,25))
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.2, hspace=0.5)
fig.suptitle('Category Ratings', fontsize=22)

ax = ax.reshape(-1)

for i in range(len(col_names)):
#    if i % cols == 0:
#        ax[i].set_ylabel('Frequency\n', fontsize=14)
    ax[i].set_title(col_names[i]+'\n', fontsize=16)
    ax[i].set_xlabel('Ratings', fontsize=14)
    ax[i].set_ylim([0,2.4])
    ax[i].set_xticks([0,1,2,3,4,5])
    sns.distplot([inst[i] for inst in data], ax=ax[i])

for i in range(n_plots, (rows*cols)):  
    fig.delaxes(ax[i])
```
